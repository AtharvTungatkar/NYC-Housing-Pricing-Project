```python
import numpy as np
import pandas as pd
import os

os.chdir('../')
```

```python
# Importing all datasets

price=pd.read_csv('datasets ready to merge/price.csv')
complaints_311=pd.read_csv('datasets ready to merge/311_merge_ready.csv')
rodent=pd.read_csv('datasets ready to merge/rodent.csv')
crime=pd.read_csv('datasets ready to merge/Crime.csv')
air_quality=pd.read_csv('datasets ready to merge/airquality.csv')
```

```python
price.drop(labels='Unnamed: 0',axis=1,inplace=True)
complaints_311.drop(labels='Unnamed: 0',axis=1,inplace=True)
rodent.drop(labels='Unnamed: 0',axis=1,inplace=True)
crime.drop(labels='Unnamed: 0',axis=1,inplace=True)
air_quality.drop(labels='Unnamed: 0',axis=1,inplace=True)
```

```python
def borough_classifier(borough):
    if borough==1:
        return 'manhattan'
    elif borough==2:
        return 'bronx'
    elif borough==3:
        return 'brooklyn'
    elif borough==4:
        return 'queens'
    elif borough==5:
        return 'staten island'
```

```python
price['BOROUGH']=price['BOROUGH'].apply(borough_classifier)
crime['BORO_NM']=crime['BORO_NM'].str.lower()
air_quality['BOROUGH']=air_quality['Borough'].str.lower()
```

```python
price.dropna(inplace=True)
complaints_311.dropna(inplace=True)
price['BOROUGH']
```

```python
rodent.columns
```

```python
merged_1=price.merge(complaints_311,how='left',left_on=['Year','Month','ZIP CODE'],right_on=['Year','Month','Incident Zip'])
merged_1.dropna(inplace=True)
```

```python
merged_2=merged_1.merge(rodent,how='left',left_on=['Year','Month','ZIP CODE'],right_on=['Year','Month','ZIP_CODE'])
merged_2.dropna(inplace=True)
```

```python
air_quality.head()
```

```python
merged_3=merged_2.merge(crime,how='left',left_on=['Year','Month','BOROUGH'],right_on=['Year', 'Month', 'BORO_NM'])
```

```python
data=merged_3.merge(air_quality,how='left',left_on=['Year','BOROUGH'],right_on=['Year', 'BOROUGH'])
```

```python
data.to_csv('modelling_data.csv')
```
